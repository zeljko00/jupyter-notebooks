```python
dim=img.size
x1=dim[0]/2-200
x2=x1+400
y1=dim[1]/5*3
y2=dim[1]
pc=img.crop((x1,y1,x2,y2))
```

```python
pc
```

```python
img.paste(im=pc, box=(0,0))
```

```python
img
```

Sve metode pozvane nad objektom slike se ne primjenjuju na fajl na disku, već samo na objekat - čuvanje promjena se radi putem save(PATH) metoda, kojom se na osnovu objekta slike kreira odgovrajući fajl an disku.
